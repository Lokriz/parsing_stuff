```python
import pandas as pd
import numpy as np
import re

pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_excel('YM_item_names_full.xlsx', index_col=0).reset_index(drop=True)
df
```

```python
df = pd.read_csv('YM_item_sample_1.csv', sep='\t')
df
```

```python
def get_value(text:str):
    try:
        a = re.search(r'\d+(?=₽)', text)
        b = a.group(0)
    except Exception:
        b = np.nan
    return b

def format_item_name(lst:list):
    return lst[2] if lst[0] in ['Б/У', 'Уценённый', 'Уценка'] \
            else (lst[1] if lst[0] in ['По рецепту'] else lst[0])
```

```python
df['new_val'] = df['value'].str.replace('\u202f', '').str.replace('\u2009', '')

df['new_val'] = df.apply(lambda x: get_value(x['new_val']) if not x['new_val'].isdigit() else x['new_val'], axis=1).astype('float')
df
```

```python
df['item_name_new'] = df.apply(lambda x: format_item_name(x['item_name'].split('\n')), axis=1)
df
```

```python
df1 = df.copy()

df1['item_name'] = df1['item_name_new']
df1['value'] = df1['new_val']
df1.drop(columns=['item_name_new', 'new_val'], inplace=True)
df1.drop_duplicates(inplace=True)

df1.to_excel('YM_item_names_formated_1.xlsx', index=False)
```

```python
df1['lvl2_href_new'] = df1.apply(lambda x: re.search(r'https://.+/\d+', x['lvl2_href']).group(0), axis=1)
```

```python
df1['lvl2_href_naked'] = df1.apply(lambda x: re.search(r'https://.+/\(?=d+)', x['lvl2_href']).group(0), axis=1)
```
