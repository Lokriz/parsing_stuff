```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import random

pd.set_option('display.max_rows', 3000)
pd.set_option('display.max_colwidth', 70)

from IPython.display import display
```

```python
def format_services(txt):
    services_reg = re.compile(r'ещё \d+ услуг', re.IGNORECASE)
    return [service.split('|')[0] for service in txt.split('; ') if not services_reg.search(service)] if txt else None
```

```python
avito_data = pd.read_excel('./item_names/avito_item_names.xlsx').fillna({'extra_services': ''})
avito_data
```

```python
avito_data['extra_services_formt'] = avito_data.apply(lambda x: format_services(x['extra_services']), axis=1)
```

```python
avito_data_frmt = avito_data.explode('extra_services_formt')
pd.concat([avito_data_frmt[['ind', 'href', 'item_name']], 
           avito_data_frmt[['ind', 'href', 'extra_services_formt']]\
                   .rename(columns={'extra_services_formt': 'item_name'}).dropna()])\
        .drop_duplicates().sort_values('ind').reset_index(drop=True)\
        .to_excel('./item_names/avito_full_item_names.xlsx', index=False)
```
